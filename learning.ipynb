#### Separate Validation

```python
print(validation_data)
```

##### Train

```python
from models import HMModel
```

```python
training_data = training_data.filter_markers(markers, keep=True)
```

```python
model = HMModel(n_components=10, n_mix=8, n_iter=315, tol=1, h_type='velpos', downsample_step=7, verbose=True)
```

```python
%%time
results_velpos = model.train(training_data)
```

##### Validate

```python
from models import PredictionResults
```

```python
results = PredictionResults()
```

```python
validation_data = validation_data.filter_markers(markers, keep=True)
```

```python
for cls in validation_data:
    for mocap in validation_data[cls]:
        pred = model.predict(mocap)[0]
        results.append(pred, cls.name)
```

```python
results.show()
```
